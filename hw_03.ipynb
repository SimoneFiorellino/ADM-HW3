```python
#define functions for pre-process the information

def download():
    nltk.download('stopwords')
    nltk.download('punkt')

def preprocessor(text: str):
    stop_words = set(stopwords.words('english')) #English stopwords
    tokens = word_tokenize(text.lower()) #tokenization
    ps = PorterStemmer() 
    #remove all characters that are not in a-zA-Z
    # I numeri non ci servono? tipo libri come 1984 non esistono? XD
    result = [ps.stem(re.sub(r'[^a-zA-Z]', '', unidecode(word))) for word in tokens if word not in stop_words and word.isalpha()]

    return result
```

```python
vocabulary = dict()
```

```python
bar = Counter('Progress: ')

for i in range(1, 301):
    path = f'E:/Universita_SAPIENZA/ADM/GitHub_HW03/TSV/list_page_{i}/'
    
    for file in os.scandir(path):
        with open(path + file.name) as tsv_file:
            tsv_reader = csv.reader(tsv_file, delimiter='\t', lineterminator='\n')
            row = list(tsv_reader)
            try:
                row = row[2][6]
                for word in preprocessor(row):
                    if word not in vocabulary.keys():
                        vocabulary[word] = len(vocabulary)
            except:
                print(f'{path + file.name}')
     
        bar.next()
bar.finish()

with open('E:/Universita_SAPIENZA/ADM/GitHub_HW03/vocabulary.json', 'w') as json_file:
    json.dump(vocabulary, json_file)

print('File generated.')
```

```python
invertedIndex = dict()

with open('E:/Universita_SAPIENZA/ADM/GitHub_HW03/vocabulary.json', 'r') as json_file:
    vocabulary = json.load(json_file)

for word in vocabulary:
    invertedIndex[vocabulary[word]] = []

bar = Counter('Progress: ')
for i in range(1, 301):
    path = f'E:/Universita_SAPIENZA/ADM/GitHub_HW03/TSV/list_page_{i}/'
    
    for file in os.scandir(path):
        with open(path + file.name) as tsv_file:
            documentId = re.sub(r'(article\_|\.tsv)', '', file.name) #i want only the number of the file
            tsv_reader = csv.reader(tsv_file, delimiter='\t', lineterminator='\n')
            row = list(tsv_reader)
            try:
                row = row[2][6]
                
                for word in preprocessor(row):
                    if documentId not in invertedIndex[vocabulary[word]]:
                        invertedIndex[vocabulary[word]].append(documentId)
            except:
                print(f'{path + file.name}')


        
        bar.next()
bar.finish()


with open('E:/Universita_SAPIENZA/ADM/GitHub_HW03/invertedIndex.json', 'w') as json_file:
    json.dump(invertedIndex, json_file)

print('File generated.')
```

```python
query = input('Query: ')
query = preprocessor(query)

with open('E:/Universita_SAPIENZA/ADM/GitHub_HW03/vocabulary.json', 'r') as json_file:
    vocabulary = json.load(json_file)

with open('E:/Universita_SAPIENZA/ADM/GitHub_HW03/invertedIndex.json', 'r') as json_file:
    invertedIndex = json.load(json_file)

documentList = []
for word in query:
    wordId = vocabulary[word]
    documentList.append(invertedIndex[str(wordId)])

if len(documentList) > 1:
    resultId = []
    for documentId in documentList[0]:
        common = True
        for j in range(1, len(documentList)):
            if documentId not in documentList[j]:
                common = False
                break
        if common:
            resultId.append(documentId)
else:
    resultId = documentList[0]

resultId = list(map(int, resultId))

for i in resultId:
    with open(f'E:/Universita_SAPIENZA/ADM/GitHub_HW03/TSV/list_page_{i // 100 + 1}/article_{i}.tsv') as tsv_file:
        tsv_reader = csv.DictReader(tsv_file, delimiter='\t')
        row = tsv_reader.__next__()

    with open('E:/Universita_SAPIENZA/ADM/GitHub_HW03/url_collected.txt') as urls:
        for j, url in enumerate(urls):
            if j == i:
                print(f"Title: {row['bookTitle']}\nUrl: {url}\nPlot:\n{row['plot']}")
                print('--------------------------------------------------')
```
